```python
import pandas as pd
df=pd.read_csv("train.csv")
df.head()
df.info()

df['Age']=df['Age'].fillna(df['Age'].median())

df=df.drop("Cabin",axis=1)

df=df.dropna(subset='Embarked')
df.isnull().sum()
```

```python
df.groupby('Sex')['Survived'].mean()
df.groupby('Pclass')['Survived'].mean()
df.groupby(["Sex","Pclass"])["Survived"].mean()
```

```python
import matplotlib.pyplot as plt
df.groupby(["Sex","Pclass"])["Survived"].mean().unstack().plot(kind='bar')
plt.title('survive rate by sex and pclass')
plt.ylabel("survive rate")
plt.show()
```

## 性别与舱位的交叉分析

将性别和舱位等级结合分析后发现差距进一步放大：一等舱女性生还率高达 96.7%，几乎全部生还；而三等舱男性生还率仅 13.5%，是所有群体中最低的。说明性别和舱位等级是两个独立且叠加的关键因素，共同决定了乘客的生还概率。

```python
df['agegroup']=pd.cut(df['Age'],bins=[0,12,18,35,60,100],labels=['Child', 'Teen', 'YoungAdult', 'MiddleAge', 'Senior'])
df['agegroup'].value_counts()
df.groupby('agegroup')["Survived"].mean()
```

```python
df.groupby('agegroup')["Survived"].mean().plot(kind='bar')
plt.title("survive rate by agegroup")
plt.ylabel("survive rate")
plt.show()
```

## 年龄分组分析

儿童生还率最高（58%），老年人最低（19%），符合"儿童优先"的逃生原则。
